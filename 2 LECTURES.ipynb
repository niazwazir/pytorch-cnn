```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
# our model for the forward pass so make a function, x is input
def forward(x):
    return x * w   #o/p y = x*w

# Loss function
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y) #loss = (y-y^)^2
```

```python
w_list = []
mse_list = []

for w in np.arange(0.0, 5, 1):  #random guess no, start 0.0 end at 4.1 and step size 0.1 
    print("w=", w) #print value of w
    l_sum = 0 #initialize loss as 0
    for x_val, y_val in zip(x_data, y_data): # call dataset and get x and y values
        y_pred_val = forward(x_val) #get prediction value
        l = loss(x_val, y_val) #compare the loss
        l_sum += l # sum this loss and printout
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum / 3) #divided by 3, because 3 values
    w_list.append(w)
    mse_list.append(l_sum / 3)
#Result shows w change mse also change
```

```python
from IPython.display import Image
Image('TrainingLoss.png')
```

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
